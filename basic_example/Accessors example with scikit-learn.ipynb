## Example of extending Pandas with the Accessor feature

Accessors are a simple and tidy way for extending the functionality of pandas, and are very useful for adding processing code and custom plotting. Below is an example of using the accessor defined in `accessor.py` on an example pandas dataframe. Styling for the plots is controlled via the matplotlib style sheet defined in `custom_stype.mplstyle`.

See this link for more details: https://pandas.pydata.org/docs/development/extending.html#subclassing-pandas-data-structures

```python
import pandas as pd
from sklearn import datasets

# Load the accessor for customising dataframe plotting
import accessor
```

#### Load a sample dataset from Scikit-Learn's collection of datasets

```python
wine = datasets.load_wine()
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df["target"] = wine.target
wine_df
```

### Using the accessor

Some examples of using the plotting available via the accessor are show below. This is only a basic example, but notice how a small watermark has been added to each plot.

```python
wine_df.custom_plot.hist("ash")
```

```python
wine_df.custom_plot.scatter("ash", "malic_acid")
```
